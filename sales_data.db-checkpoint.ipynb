### Import Required Libraries

```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### Connect to the SQLite Database

```python
conn = sqlite3.connect("sales_data.db")
cursor = conn.cursor()
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS movie_sales (
    movie_title TEXT,
    quantity INTEGER,
    price REAL
)
""")
```

```python
sample_data = [
    ("Avengers", 120, 10),
    ("Avengers", 80, 10),
    ("Inception", 50, 8),
    ("Titanic", 200, 7),
    ("Inception", 70, 8)
]
```

```python
cursor.execute("DELETE FROM movie_sales")  # Clear old data for fresh run
cursor.executemany("INSERT INTO movie_sales VALUES (?, ?, ?)", sample_data)
conn.commit()
```

```python
query = """
SELECT movie_title, 
       SUM(quantity) AS total_tickets_sold,
       SUM(quantity * price) AS total_revenue
FROM movie_sales
GROUP BY movie_title
ORDER BY SUM(quantity * price) DESC;
"""

df = pd.read_sql_query(query, conn)
```

```python
print("=== Movie Sales Summary ===")
print(df)
```

### Plot Basic Bar Chart

```python
plt.figure(figsize=(6, 4))
plt.bar(df['movie_title'], df['total_revenue'], color='skyblue')
plt.xlabel("Movie Title")
plt.ylabel("Total Revenue")
plt.title("Movie Revenue Summary")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(3, 3))
plt.pie(df['total_revenue'], labels=df['movie_title'], autopct='%1.1f%%', startangle=140)
plt.title("Revenue Share by Movie")
plt.tight_layout()
plt.show()
```

```python
x = np.arange(len(df['movie_title']))  
bar_width = 0.35

plt.figure(figsize=(5, 3))
plt.bar(x - bar_width/2, df['total_tickets_sold'], width=bar_width, label='Tickets Sold', color='skyblue')
plt.bar(x + bar_width/2, df['total_revenue'], width=bar_width, label='Revenue', color='orange')

plt.xticks(x, df['movie_title'])
plt.xlabel("Movie Title")
plt.ylabel("Value")
plt.title("Tickets Sold vs Revenue")
plt.legend()
plt.tight_layout()
plt.show()
```
